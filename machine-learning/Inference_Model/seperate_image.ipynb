```python
from yolo.YoloInferenceModel import YoloInferenceModel
import os
import cv2
import json
```

```python
YOLO_MODEL_DIR = os.path.join('./trained_models', 'YOLO')
YOLO_BEST_MODEL = os.path.join('model_train_renfred_1', 'weights', 'best.pt')
YOLO_MODEL_PATH = os.path.join(YOLO_MODEL_DIR, YOLO_BEST_MODEL)

extra_dataset = './extra_dataset/train/'
clean_extra_ds = './extra_dataset_clean/train'
```

```python
directory = os.path.join("temp_image/extra_dataset_2/")

print(len(os.listdir(directory)))
```

```python
result = {}

yolo_model = YoloInferenceModel(YOLO_MODEL_PATH)

os.makedirs(extra_dataset, exist_ok=True)

ndjson_path = os.path.join(extra_dataset, 'result.ndjson')


for file_index, file in enumerate(os.listdir(directory)):
    file_index += 302
    file_name, extension = os.path.splitext(os.path.basename(file))
    file_path = os.path.join(directory, file)
    
    # Read the image
    image = cv2.imread(file_path)
    if image is None:
        print(f"Error: Could not read image {file_path}")
        continue

    # Process the image using YOLO model
    cropped_image = yolo_model.process_image(image)
    for class_name, images_data in cropped_image.items():
        if 'receipt' in class_name:
            print("Skipped Receipt class")
            continue

        for _, data in enumerate(images_data):
            cropped_img = data['image']
            new_file_name = f'{file_index}_{class_name}.jpg'
            save_path = os.path.join(extra_dataset, new_file_name)
            
            # Save the cropped 
            cv2.imwrite(save_path, cropped_img)
        
    # Open the NDJSON file for writing

with open(ndjson_path, 'w') as ndjson_file:
        for file in sorted(os.listdir(extra_dataset)):
            json_record = {
                    "file_name": file,
                    "text": ""
                }
            ndjson_file.write(json.dumps(json_record) + '\n')

print(f"Results saved to {ndjson_path}")
```

```python
def load_json_data(file_path):
    image_data = []
    text_data = []
    with open(file_path, 'r') as file:
        for line in file:
            try:
                data = json.loads(line.strip())
                image_data.append(data['file_name'])
                text_data.append(data['text'])
            except json.JSONDecodeError as e:
                print(f"Error decoding JSON: {e}")
    
    return image_data, text_data
```
